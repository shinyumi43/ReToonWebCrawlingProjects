```python
col1='genre' # 장르
col2='title' # 제목
col3='writer' # 작가
col4='starRate' # 평점
col5='imgUrl' # 이미지 url
col6='introduce' # 작품 소개
col7='genreDetails' # 세부 장르
col8='ageCriteria' # 연령 기준
col9='serialInformation' # 연재 정보
col10='platform' #플랫폼 정보
col_list=[col1, col2, col3, col4, col5, col6, col7, col8, col9, col10]
df_ta=pd.DataFrame(columns=col_list)
df_ta
```

```python
# 파일을 저장할 주소
address=str('C:\\Users\\shinyumi\\Desktop\\Image\\Naver')
# 장르 목차
genre_lists=driver.find_elements(By.CSS_SELECTOR, '#content > div.snb > ul > li')
# 일상을 시작으로 추출
for j in range(5, len(genre_lists)):
    # 목록 순차적으로 이동
    driver.find_element(By.XPATH,'/html/body/div/div[3]/div[1]/div[1]/ul/li['+str(j+1)+']/a').click()
    
    # 장르 추출
    genre=driver.find_element(By.XPATH,'/html/body/div/div[3]/div[1]/div[1]/ul/li['+str(j+1)+']/a').text
    
    # 목록에 따른 웹툰 개수를 추출
    webtoon_lists=driver.find_elements(By.CSS_SELECTOR, '#content > div.list_area > ul > li')
    # 웹툰 개수만큼 정보를 추출
    for i in range(0, len(webtoon_lists)):
        # 웹툰 평점 추출
        starRate=driver.find_element(By.XPATH,'/html/body/div/div[3]/div[1]/div[3]/ul/li['+str(i+1)+']/dl/dd[3]/div/strong').text
        
        # 연재인지 완결인지에 대한 정보 추출
        serialInformation=""
        img_lists=driver.find_elements(By.CSS_SELECTOR, '#content > div.list_area > ul > li:nth-child('+str(i+1)+') > div > a > img')
        if len(img_lists)==2:
            finish=driver.find_element(By.XPATH, '/html/body/div/div[3]/div[1]/div[3]/ul/li['+str(i+1)+']/div/a/img[2]').get_attribute("alt")
            serialInformation=str(finish)
        else:
            serialInformation="연재"
        
        # 웹툰 세부 내용 페이지로 이동
        driver.find_element(By.XPATH,'/html/body/div/div[3]/div[1]/div[3]/ul/li['+str(i+1)+']/dl/dt/a').click()
        # 이동하면서 웹페이지 로드되는 동안 기다림
        time.sleep(2)
        
        # 웹툰 제목 추출
        title=driver.find_element(By.CSS_SELECTOR, '#content > div.comicinfo > div.detail > h2 > span.title').text
        # 웹툰 작가 추출
        writer=driver.find_element(By.CSS_SELECTOR, '#content > div.comicinfo > div.detail > h2 > span.wrt_nm').text
        writer=writer.replace(' / ', ', ') # 가공
        
        # 이미지 url 추출
        imgUrl = driver.find_element(By.CSS_SELECTOR,'#content > div.comicinfo > div.thumb > a > img').get_attribute("src")
        
        # 파일 저장 시에 사용할 수 없는 문자가 존재하므로 해당 문자는 삭제
        result=title
        for char in title:
            if char in "\/:*?""<>|":
                result=result.replace(char, '')
        
        # 해당 주소에 웹툰 제목으로 파일을 저장
        urllib.request.urlretrieve(imgUrl,address+"/"+str(result)+".jpg")
        # 저장되는 시간을 설정
        #time.sleep(2)
        # 저장된 이미지 받아오기
        #img = Image.open(address+"/"+str(title)+".jpg")
        # 이미지 크기 조정
        #after_img = img.resize((128, 192))
        # 이미지 저장
        #after_img.save(address+"/"+str(title)+".jpg")
        
        # 웹툰 소개글 추출
        introduce=driver.find_element(By.CSS_SELECTOR,'#content > div.comicinfo > div.detail > p:nth-child(2)').text
        
        # 연령 기준이 존재하지 않는 경우로 인해 리스트 형식으로 반환 후에 저장
        elements=driver.find_elements(By.CSS_SELECTOR, '#content > div.comicinfo > div.detail > p.detail_info > span')
        result_elements=[]
        for elem in elements:
            result_elements.append(elem.text)
        # 세부 장르와 연령 기준 모두 존재하는 경우
        if len(result_elements)==2:
            genreDetails=result_elements[0]
            ageCriteria=result_elements[1]
        # 세부 장르는 존재하지만, 연령 기준이 존재하지 않는 경우
        elif len(result_elements)==1:
            genreDetails=result_elements[0]
            ageCriteria=''
        
        # 세부 장르 내용 추출
        # genreDetails=driver.find_element(By.CSS_SELECTOR,'#content > div.comicinfo > div.detail > p.detail_info > span.genre').text
        # 연령 기준 추출
        # ageCriteria=driver.find_element(By.CSS_SELECTOR,'#content > div.comicinfo > div.detail > p.detail_info > span.age').text
        
        # 웹툰 정보에 따른 단일 배열 설정
        df_tem={col1:genre, col2:title, col3:writer, col4:starRate, col5:imgUrl, col6:introduce, col7:genreDetails, col8:ageCriteria, col9:serialInformation, col10:'네이버웹툰'}
        
        # 최종 데이터 프레임에 삽입
        df_ta=df_ta.append(df_tem, ignore_index=True)
        
        # 로그 기록
        print(str(title)+' 추출')
        
        # 뒤로 가기
        driver.back()
        time.sleep(2)
        
# 최종 데이터 프레임 출력
df_ta
```

```python
# 저장할 엑셀의 경로
excel_path='C:\\Users\\shinyumi\\Desktop\\Excel\\네이버웹툰.xlsx'
# 엑셀에 작성하기, 파일이름만 변경해서 저장
writer = pd.ExcelWriter(excel_path, engine='xlsxwriter')
df_ta.to_excel(writer, index=False)
writer.save()
```

```python
# 파일 내에 모든 사진을 한 번에 변경하기 위한 부분
import os
import glob
from PIL import Image
```

```python
# 파일을 저장할 주소
address=str('C:\\Users\\shinyumi\\Desktop\\Image\\Naver')
# 일단 원본 이미지로 저장한 뒤에 프론트엔드로 진행해야할 수도 있고, 크롤링을 통해서 규격에 맞추어 전달해야할 수 있음
files = glob.glob(address+'/*')

for f in files:
    try:
        # 이미지 반환
        img = Image.open(f)
        # 이미지 크기 조정
        after_img = img.resize((128, 192))
        # 이미지 파일 제목과 확장자 반환
        name, ext = os.path.splitext(f)
        # 이미지 저장
        after_img.save(name+ext)
    except OSError as e:
        pass
```
